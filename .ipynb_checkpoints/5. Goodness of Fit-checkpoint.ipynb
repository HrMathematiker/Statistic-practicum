# №1

Закон Бенфорда в его житейской интерпретации утверждает, что если выбрать набор каких-то чисел из реальных данных, то распределение первых цифр этих чисел будет иметь вид 
$$P(\{k - \text{первая цифра}\}) = \log_{10}(1 + \frac{1}{k}),\text{ k = 1, ..., 9}.$$ 

Проверить его 

а) на списке стран по населению, 

б) на числах Фибоначчи

```python
population_df = pd.read_csv('population_by_country_2020.csv')
population_df.head()
```

```python
def chi2_benford_law(data: list):
    first_digit_list = [int(str(num)[:1]) for num in data]
    benford_dist = [np.log10(1 + 1/k) for k in range(1, 10)]
    benford_dist = np.array(benford_dist)
    first_digit_frequencies = Counter(first_digit_list)
    first_digit_frequencies = dict(first_digit_frequencies)
    first_digit_frequencies = [first_digit_frequencies[num] 
                               for num in range(1, 10)]
    hypothetic_frequencies = sum(first_digit_frequencies) * benford_dist
    hypothetic_frequencies = np.round(hypothetic_frequencies)
    chi2_result = chisquare(first_digit_frequencies, 
                            hypothetic_frequencies)
    return chi2_result
```

```python
population_ls = population_df['Population (2020)']
poplation_benford_result = chi2_benford_law(population_ls)
print(f"Chi-square p-value for population data is {poplation_benford_result.pvalue:.3f}")
```

```python
fibonacci_array = [0, 1]


def fibonacci(n: int): 
    if n < 0: 
        print("Incorrect input") 
    elif n <= len(fibonacci_array): 
        return fibonacci_array[n-1] 
    else: 
        temp_fib = fibonacci(n-1) + fibonacci(n-2) 
        fibonacci_array.append(temp_fib) 
        return temp_fib
```

```python
fibonacci(100)
```

```python
fibonacci_benford_result = chi2_benford_law(fibonacci_array)
print(f"Chi-square p-value for Fibonacci numbers is {fibonacci_benford_result.pvalue:.3f}")
```

# №2

В файле Rainfall.txt приведены данные о рекордах выпадения дождя в австралийском местечке под названием Туррамура. 

Можно ли утверждать, что они подчиняются 

а) экспоненциальному 

б) логнормальному распределению?

В обоих случаях использовать и визуальный метод проверки.

```python
rainfall_data = list()
with open('Rainfall.txt', 'r') as file:
    for row in file:
        row = float(row)
        rainfall_data.append(row)
```

```python
data_mean = np.mean(rainfall_data)
exp_sample = np.random.exponential(scale=data_mean, size=10000)
lognormal_mu = np.mean(np.log(rainfall_data))
lognormal_sigma = np.std(np.log(rainfall_data))
lognormal_sample = np.random.lognormal(mean=lognormal_mu, sigma=lognormal_sigma,
                                       size=10000)
```

```python
plt.figure(figsize=(10, 10))

sns.kdeplot(data=rainfall_data, shade=True, color='b', alpha=0.3)
sns.kdeplot(data=lognormal_sample, alpha=1, color='green')
sns.kdeplot(data=exp_sample, alpha=1, color='orange')

blue_patch = mpatches.Patch(color='blue', label='rainfall data')
orange_patch = mpatches.Patch(color='orange', label='exponential pdf')
green_patch = mpatches.Patch(color='green', label='lognormal pdf')

plt.legend(loc='upper right', handles=[blue_patch, orange_patch, green_patch])
```

```python
ad_test_expon = ad_test(rainfall_data, expon(scale=data_mean))
ks_test_expon = ks_test(rainfall_data, expon(scale=data_mean))
cvm_test_expon = cvm_test(rainfall_data, expon(scale=data_mean))
ad_test_lognorm = ad_test(rainfall_data, 
                          lognorm(s=lognormal_sigma, scale=np.exp(lognormal_mu)))
ks_test_lognorm = ks_test(rainfall_data, 
                          lognorm(s=lognormal_sigma, scale=np.exp(lognormal_mu)))
cvm_test_lognorm = cvm_test(rainfall_data, 
                            lognorm(s=lognormal_sigma, scale=np.exp(lognormal_mu)))
print(f'Anderson-Darling p-value of exponential hypothesis: {ad_test_expon.pvalue:.3f}')
print(f'Kolmogorov-Smirnov p-value of exponential hypothesis: {ks_test_expon.pvalue:.3f}')
print(f'Cramer-von Mises p-value of exponential hypothesis: {cvm_test_expon.pvalue:.3f}\n')
print(f'Anderson-Darling p-value of lognormal hypothesis: {ad_test_lognorm.pvalue:.3f}')
print(f'Kolmogorov-Smirnov p-value of lognormal hypothesis: {ks_test_lognorm.pvalue:.3f}')
print(f'Cramer-von Mises p-value of lognormal hypothesis: {cvm_test_lognorm.pvalue:.3f}')
```

```python
pylab.figure(figsize=(10, 10))
probplot(rainfall_data, dist=expon(scale=data_mean), plot=pylab)
pylab.title(label='Exponential distribution quantile-quantile plot')
pylab.show()
```

```python
pylab.figure(figsize=(10, 10))
probplot(rainfall_data, dist=lognorm(s=lognormal_sigma, scale=np.exp(lognormal_mu)), plot=pylab)
pylab.title(label='Lognormal distribution quantile-quantile plot')
pylab.show()
```
