```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
```

```python
#2
```

```python
def norm_check(estimation_ls, param):
    normalizator = lambda x: np.sqrt(len(estimation_ls)) * (x - param)
    return list(map(normalizator, estimation_ls))
```

```python
uniform_MLE = lambda x: np.max(np.abs(x))
```

```python
def MPS(sample):
    max_sm = np.max(sample)
    min_sm = np.min(sample)
    n = len(sample)
    discr = ((n * (min_sm + max_sm)) ** 2 - 4 * min_sm * max_sm)
    estimation = (n * (max_sm - min_sm) + np.sqrt(discr)) / (2 * (n - 1))
    return estimation
```

```python
theta = 100
sample_size = 500
num_of_samples = 10000
uniform_sample = np.random.uniform(-theta, theta, size=(num_of_samples, sample_size))
```

```python
MLE_list = list(map(uniform_MLE, uniform_sample))
sns.kdeplot(data=MLE_list, shade=True,  label='uniform MLE')
```

```python
normalized_MLE = norm_check(MLE_list, theta)
sns.distplot(normalized_MLE, label='normalized uniform MLE')
```

```python
MPS_list = list(map(MPS, uniform_sample))
sns.distplot(MPS_list, label='uniform MPS')
```

```python
normalized_MPS = norm_check(MPS_list, theta)
sns.kdeplot(data=normalized_MPS, shade=True, label='normalized uniform MPS')
```

```python
#parameters must be above zero
a = 10
b = 20 
sample_size = 5000
num_of_samples = 1000
```

```python
beta_samples = np.random.beta(a, b, size=(num_of_samples, sample_size))
```

```python
def beta_likelihood(param, *args):
    a, b = param
    data = args[0]
    pdf = beta.pdf(data, a, b, loc=0, scale=1)
    log = np.log(pdf)
    log = np.where(log == -np.inf, 0, log)
    likelihood = -1 * np.sum(log)
    return likelihood
```

```python
alpha_MLE_list = []
beta_MLE_list = []
bounds = ((0, np.inf), (0, np.inf))
for sample in beta_samples:
    result = minimize(beta_likelihood, x0=[0.1, 0.1], args=sample, method='Nelder-Mead', bounds=bnds)
    alpha_MLE_list.append(result.x[0])
    beta_MLE_list.append(result.x[1])
```

```python
len(alpha_MLE_list)
```

```python
sns.kdeplot(data=alpha_MLE_list, shade=True, label='alpha MLE')
```

```python
sns.kdeplot(data=beta_MLE_list, shade=True, label='beta MLE')
```

```python
normalized_alpha_MLE = norm_check(alpha_MLE_list, a)
normalized_beta_MLE = norm_check(beta_MLE_list, b)
```

```python
sns.kdeplot(data=normalized_alpha_MLE, shade=True, label='normalized alpha MLE')
```

```python
sns.kdeplot(data=normalized_beta_MLE, shade=True, label='normalized beta MLE')
```
