```python
import numpy as np
from scipy.stats import expon, foldnorm, bernoulli, uniform
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
import random
```

```python
#1
```

```python
def sampleInCircle(sampleSize, radius=1):
    angleList = uniform.rvs(0, 2 * np.pi, size=sampleSize)
    x_list = [radius * np.cos(angle) for angle in angleList]
    y_list = [radius * np.sin(angle) for angle in angleList]
    return list(zip(x_list, y_list))
```

```python
numOfSamples = 200
listOfSamples = sampleInCircle(numOfSamples)
x = [point[0] for point in listOfSamples]
y = [point[1] for point in listOfSamples]
plt.figure(figsize=(8,8))
plt.scatter(x, y)
```

```python
ExpectationEstimation = sum(np.abs(x)) / len(x)
print('Estimation of |x| expectation: {:.3f}'.format(ExpectationEstimation))
```

```python
#с = 0, loc = 0, scale = 1
bnds = (0, np.inf)
negativeRatioPdf = lambda x: -1 * foldnorm.pdf(x, c=0, scale=1) / expon.pdf(x)
maxOfRatio = minimize(negativeRatioPdf, x0=[0.1], method='Nelder-Mead', bounds=bnds)
```

```python
c = -1 * maxOfRatio.fun
probFunction = lambda x: foldnorm.pdf(x, c=0, scale=1) / (c * expon.pdf(x))
```

```python
numOfSamples = 100
foldedNormSamples = []
while len(foldedNormSamples) < numOfSamples:
    exponSample = expon.rvs()
    curProbability = probFunction(exponSample)
    bernSample = bernoulli.rvs(curProbability)
    if bernSample == 1:
        foldedNormSamples.append(exponSample)
```

```python
sns.distplot(foldedNormSamples, label='pdf of folded norm dist')
```

```python
#делаем стандартные нормальные с.в. 
positiveOrNegativeArray = np.array(random.choices([-1, 1], k=numOfSamples))
normDistrSamples = positiveOrNegativeArray  * np.array(foldedNormSamples)
```

```python
sns.distplot(normDistrSamples)
```

```python
def mapGenerator(mapWidth=100, mapHeight=100, numOfCities=5, numOfMines=2):
    generatedMap = np.zeros((mapWidth, mapHeight))
    AvailablePointsForCities = [np.array(point[0]) for point in np.ndenumerate(generatedMap)]
    AvailablePointsForMines = np.array(AvailablePointsForCities)
    for _ in range(numOfCities):
        if len(AvailablePointsForCities) == 0:
            raise Exception('Generation is impossible')
        newCity = random.choice(AvailablePointsForCities)
        generatedMap[newCity[0], newCity[1]] = 1
        PointsForMinesInNewCity = [point for point in AvailablePointsForMines 
                                   if 10 <= np.linalg.norm(point - newCity) <= 20]
        
        for _ in range(numOfMines):
            if len(PointsForMinesInNewCity) == 0:
                raise Exception('Generation is impossible')
            newMine = random.choice(PointsForMinesInNewCity)
            generatedMap[newMine[0], newMine[1]] = 2
            PointsForMinesInNewCity = [point for point in PointsForMinesInNewCity 
                                       if np.linalg.norm(point - newMine) >= 10]
            AvailablePointsForMines = [point for point in  AvailablePointsForMines 
                                        if np.linalg.norm(point - newMine) >= 10]
        AvailablePointsForCities = [point for point in AvailablePointsForCities 
                                    if np.linalg.norm(point - newCity) >= 40]
    return generatedMap
```

```python
#синее - город, желтое - шахта
newMap = mapGenerator()
plt.figure(figsize=(10, 10))
plt.imshow(newMap)
```
